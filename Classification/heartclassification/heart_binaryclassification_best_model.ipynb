#Categorical varaibles to label encoding and one hot encoding


```python
from sklearn.preprocessing import LabelEncoder

dataframe['Sex_end'] = LabelEncoder().fit_transform(dataframe['Sex'])
dataframe['ExerciseAngina_end'] = LabelEncoder().fit_transform(dataframe['ExerciseAngina'])
```

```python
dataframe.drop(['Sex','ExerciseAngina'],axis=1,inplace=True)
```

```python
dataframe= pd.get_dummies(dataframe,columns=['ChestPainType','RestingECG','ST_Slope'])
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
#Scale the data
scaler = StandardScaler()
scaler.fit(dataframe)
scaled = scaler.transform(dataframe)

#Obtain principal components
pca = PCA().fit(scaled)

pc = pca.transform(scaled)
pc1 = pc[:,0]
pc2 = pc[:,1]

#Plot principal components
plt.figure(figsize=(10,10))

colour = ['#ff2121' if y == 1 else '#2176ff' for y in dataframe['HeartDisease']]
plt.scatter(pc1,pc2 ,c=colour,edgecolors='#000000')

plt.yticks(size=12)
plt.xticks(size=12)
plt.xlabel('PC1')
plt.ylabel('PC2')
```

# Applying different Binary classification models

```python
from datetime import datetime
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB

from sklearn.model_selection import GridSearchCV
```

```python
model_params= {
    'logistic_regression':{'model': LogisticRegression(),
                           'params':{
                               'max_iter':[400,600,800,1000,1200],
                               'solver': [ 'lbfgs', 'liblinear'],
                               'C':[1.0,0.1,0.01,0.001]
                           }
                        },
    'knn':{'model': KNeighborsClassifier(),
                           'params':{
                               'n_neighbors':[1,2,3,4,5,6,7,8,9]

                           }
        },
    'svm':{'model': SVC(),
                           'params':{
                               'kernel':['linear','sigmoid','poly','rbf']

                           }
        },
    'decision_tree_classifier':{'model': DecisionTreeClassifier(),
                                'params':{
                                    'max_depth':[50,80,100,None]
                                }
                                
                               },
    'bagging_classifier':{'model':BaggingClassifier(DecisionTreeClassifier()),
                          'params':{
                              'max_samples':[0.3,0.5,0.7,1],
                              'max_features':[0.5,0.8,1],
                              'n_estimators':[5,10,15,20,25,30,35]
                              
                          }
                         },
                          
        
    
    'adaboost_classifier':{
        'model': AdaBoostClassifier(DecisionTreeClassifier(min_samples_split=10,max_depth=4)),
        'params': {
            'n_estimators':[5,10,15,20,25,30],
            'learning_rate':[0.3,0.5,0.7,1]
        }
    },
    
    'random_forest':{
        'model':RandomForestClassifier(),
        'params':{
            'n_estimators': [20,30,40,50],
            'max_depth':[5,10,15,20],
            'min_samples_split':[5,10]
        }
    }
    
    
}
```

```python
scores=[]
for model_name,mp in model_params.items():
    clf= GridSearchCV(mp['model'],mp['params'],cv=3,return_train_score=False)
    clf.fit(X_train,Y_train)
    scores.append({
        'model':model_name,
        'best_score':clf.best_score_,
        'best_params': clf.best_params_
    })
```

```python
df= pd.DataFrame(scores,columns=['model','best_score','best_params'])
df
```
